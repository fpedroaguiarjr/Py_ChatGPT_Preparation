```python
#IMPORTANDO A BIBLIOTECA PANDAS
import pandas as pd
```

```python
# LENDO O DATASET E MOSTRANDO NA TELA
df = pd.read_excel('data.xlsx')
display(df)
```

```python
# VERIFICANDO OS TIPOS DE FORMATOS DAS COLUNAS
df.dtypes # Importante observar que alguns tipos estão com seu formato inadequado (Ex: CILINDROS, VALOR
# ...estão como "object", quando o certo seria em formato numérico como "float" ou "int")
```

```python
# TRANFORMANDO AS COLUNAS PARA FORMATOS ADEQUADOS
df_converter_colunas = ['KM-LITRO', 'CILINDROS', 'PESO(KG)', 'VEL_MAXIMA(KM/H)',
                          'CAVALOS', 'VALOR']
df[df_converter_colunas] = df[df_converter_colunas].apply(pd.to_numeric, errors = 'coerce')
df.dtypes # As colunas foram tranformadas para "float64"
```

```python
# APAGANDO AS COLUNAS DUPLICADAS
# df.drop_duplicates(inplace = True) --> Como não temos colunas duplicadas, verificamos as linhas
# APAGANDO LINHAS DUPLICADAS
df = df.drop_duplicates(['NOME']) # 
display(df)
```

```python
# PREENCHENDO DADOS FALTANTES / EXCLUINDO DADOS FALTANTES
df = df.dropna()
display(df)
```

```python
# NORMALIZANDO AS "STRINGS" NO DATASET
df['E-MAIL'] = df['E-MAIL'].str.replace("gmial", "gmail")
colunms_to_low = ['NOME','SOBRENOME', 'E-MAIL', 'MODELO','TRANSMISSAO', 'CATEGORIA']
for col in colunms_to_low:
    df[col] = df[col].str.lower()
display(df)    
```
